```python
import fitz  # PyMuPDF
import os
from PIL import Image
import io

# Create folders to store output
os.makedirs("extracted_images", exist_ok=True)
os.makedirs("case_studies", exist_ok=True)

pdf_path = "Case_study_1.pdf"
doc = fitz.open(pdf_path)

case_studies = []

for page_num, page in enumerate(doc):
    text = page.get_text("text")

    # Save the text as a separate .txt file (optional)
    with open(f"case_studies/page_{page_num + 1}.txt", "w", encoding='utf-8') as f:
        f.write(text)

    # Extract and save images
    for img_index, img in enumerate(page.get_images(full=True)):
        xref = img[0]
        base_image = doc.extract_image(xref)
        image_bytes = base_image["image"]
        image = Image.open(io.BytesIO(image_bytes))
        image_path = f"extracted_images/page_{page_num + 1}_img_{img_index + 1}.png"
        image.save(image_path)

    case_studies.append({
        "page_number": page_num + 1,
        "text": text,
        "images": [f"extracted_images/page_{page_num + 1}_img_{i + 1}.png" for i in range(len(page.get_images(full=True)))]
    })

doc.close()
```

```python
import pandas as pd

df = pd.DataFrame(case_studies)
df.head()
```

```python
def clean_text(text):
    return ' '.join(text.replace('\n', ' ').split())

df['cleaned_text'] = df['text'].apply(clean_text)
```
